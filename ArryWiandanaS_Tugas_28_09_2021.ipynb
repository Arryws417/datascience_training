```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jumadi-cloud/Fundamental-Python/main/Dataset/konsumen2.csv')
df.keys()
```

```python
my_data = pd.DataFrame(df)
my_data.head()
```

```python
x = np.array(df)
print(x)
```

```python
plt.scatter(x[:,0],x[:,1], label='True position')
plt.xlabel("gaji")
plt.ylabel("pengeluaran")
plt.title("grafik pengeluaran")
plt.show()
```

```python
#menggunakan library kmeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(x[:,0],x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("gaji")
plt.ylabel("pengeluaran")
plt.title("grafik konsumen")
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='green')
plt.xlabel("gaji")
plt.ylabel("pengeluaran")
plt.title("grafik pengeluaran")
```
