PROGRAM MENGHITUNG LUAS DAN KELILING

```python
print("***Menghitung Luas & Keliling Bangun Datar***")
print("1. Persegi ")
print("2. Lingkaran ")
print("3. Persegi Panjang ")
choice=int(input("Pilih : "))
if choice == 1:
  sisi=int(input("Masukan Sisi : "))
  luas_persegi= sisi*sisi
  keliling_persegi = 4*sisi
  print("Luas Persegi adalah : ",luas_persegi)
  print("keliling Persegi adalah : ",keliling_persegi)
elif choice == 2:
  jari_jari=float(input("Masukan Jari-Jari : "))
  phi = 3.14;
  luas_lingkaran= phi*jari_jari*jari_jari;
  keliling = phi*2*jari_jari;
  print("Luas Lingkaran adalah : ",luas_lingkaran, "cm^2")
  print("Keliling Lingkaran adalah : ",keliling,"cm^2")
elif choice == 3:
  panjang=int(input("Masukan Panjang : "))
  lebar=int(input("Masukan Lebar : "))
  luas_persegipanjang = panjang*lebar;
  keliling_persegipanjang = 2*panjang+lebar;
  print("Luas Persegi Panjang : ",luas_persegipanjang,"cm^2")
  print("Kelilng Persegi Panjang : ",keliling_persegipanjang,"cm^2")
else:
  print("Pilihan Tidak ada !!")
```

KONVERSI DESIMAL KE HEXA, BINER DAN OKTAL


```python
print("***Menghitung Konversi Bilangan Desimal Ke Oktal Hex, dan Biner***")
number = int(input("Masukan Bilangan Desimal (10) : "))
print("1. Desimal Ke Hex ")
print("2. Desimal Ke Oktal ")
print("3. Desimal Ke Biner ")
choice_2 = int(input("Pilih Konversi : "))
if choice_2 == 1:
  print("Bilangan Desimal " +str(number) +" ke Hex Adalah " + hex(number).lstrip("0x").rstrip("L"))
elif choice_2 == 2:
  print("Bilangan Desimal " +str(number) + " ke Oktal Adalah "+oct(number).lstrip("0o").rstrip("L"))
elif choice_2 == 3:
  print("Bilangan Desimal " +str(number) +  " ke Biner ke Biner Adalah "+bin(number).lstrip("0b").rstrip("L"))
elif choice_2 >= 3:
  print("Pilihan Tidak ada !!")
else:
  print("Thank U !!")
```
